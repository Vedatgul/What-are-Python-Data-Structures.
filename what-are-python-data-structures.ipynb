# What are Python Data Structures

Python data types have a static data type. That is, while defining a variable, no numerical, text or list definitions are made. Data types are divided into Primitive and Non-Primitive. Primitive data types are divided into 4 as int, float, string and boolean. Non-Primitive data types include List, Dictionary, Tuple, and Set.






# Primitive data structures 

**Integer:**

```python
# It is the most used data type. Used to hold medium-length numbers.

a = 5
type(a)
```

**Float:**

```python
# Represents decimal numbers

a = 5.4
type(a)
```

**String:**



```python
# Represents textual expressions. The string expression to be used is specified between two apostrophes or quotation marks.

a = 'Hello_World'
type(a)
```

**Boolean**



```python
# The bool data type is a logical data type. It can take a true or false value. It can be used for the correctness of an expression or for checking a loop. In the code below, it will return "TRUE" because the condition in bool is not met.

a = 'Hello_World'
bool(a == 'Hi_World')
```

# Non-Primitive data structures

# *List*

Lists are shown in square brackets [ ]. Also, the elements are separated by commas. List features are as follows.

* Inclusive
* It is sequential. Index operations can be done
* Changeable

```python
# Inclusive

list = ['a', 'b', 1, 2, True, [4, 5, 6]] 
type(list)
```

```python
# It is sequential. Index operations can be done

list = ['a', 'b', 1, 2, True, [4, 5, 6]] 
list[1]
```

```python
# Changeable

list = ['a', 'b', 1, 2, True, [4, 5, 6]] 
list[1]= 'c'
print(list)
```

# *Dictionary*


Dictionarys are shown in curly braces { }. Each element pair contains a key and a value. Element pairs are separated by commas. List features are as follows.

* Changeable
* Unordered. (Ordered after 3.7.)
* Inclusive.
* Key-Value




**The value part receives different types of data. So the dictionary is inclusive.**

# *Tuple*

Tuples are shown in braces ( ). Items are separated by commas. Its elements cannot be directly deleted or modified. Tuple features are as follows.

* Cannot be changed
* Sequential
* Inclusive


```python
t = ("vedat", "ahmet", 1, 2)
type(t)
```

# *Set*

Sets are represented in curly braces { } like dictionaries. Sets are used to store multiple items in a single variable. Sets do not have index and order. Set features are as follows.

*  Changeable
*  Unordered + Equal
*  It is inclusive


```python
s = {1, 5, 7, "a", "b"}
type(s)
```
